```python
import pandas as pd
from sklearn import datasets as ds 

irisDS = ds.load_iris()
iris = pd.DataFrame(irisDS.data, columns=irisDS.feature_names)
iris.columns = ["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]
iris["Series"] = "iris-"+pd.Series(irisDS["target"]).apply(lambda x: irisDS.target_names[x])
```

```python
import numpy as np

mean = np.mean(iris["SepalLengthCm"])
iris["SepalLengthCm"] = iris["SepalLengthCm"].fillna(mean)
```

```python
iris.dropna(axis=0) #drop the rows where at least one element is missing

iris.dropna(axis="columns") #drop the columns where at least one element is missing
```
