```python
#Q4 -- assign years to each of the columns, for use in aggregation
investCSV["Years"] = pd.to_datetime(investCSV["activity_date"]).dt.year
```

```python
aggByYear = investCSV.groupby(["facility_city","facility_state","Years"])["score"].mean()
aggByYear = pd.DataFrame(aggByYear.reset_index())
aggByYear.head(10)
```

```python
#find the values that need to be imputed for each city; use this approach because do not want to use mean of means to impute
#since this will be misleading
imputeValues = aggByYear.groupby(["facility_city","facility_state"])["score"].mean()
imputeValues = imputeValues.reset_index()
imputeValues.head(20)
```

```python
imputeValues.loc[imputeValues["facility_city"] == "ACTON"]["score"]
```

```python
#pivot the data so that data for each city in each row
ptCities = pd.pivot_table(aggByYear,values=["score"],index=["facility_city","facility_state"],columns=["Years"])
ptCities = pd.DataFrame(ptCities).reset_index()
ptCities.head(20)
```

```python
# need to impute the values from a previous step into the pt
ptCities["score"] = ptCities.apply(lambda x: x["score"].fillna(imputeValues.loc[imputeValues["facility_city"]==x["facility_city"].values[0,]]["score"].values[0,]),axis=1)
ptCities.head(20)

#temp = ptCities.reset_index().head(20)
#tempImpute = imputeValues[0:20]
#applymap(lambda x: print(x),level="facility_city")
#[0:5][:]["facility_city"]

#pd.DataFrame(tempImpute["score"])
#temp["score"].fillna(pd.DataFrame(tempImpute["score"]))
# this kind of works

                                   #.fillna(tempImpute.loc[tempImpute["facility_city"] == x["facility_city"]])),axis=1)
#temp.head(20)
#print(tempImpute.loc[tempImpute["facility_city"]=="ACTON"]["score"].values[0,])
```

```python
# Q5 -- % of restaurants that improve with self inspections
#find restaurants that have had a self inspections
investCSV.loc[investCSV["service_code"] == 401].head()
selfInspectOwners = investCSV.loc[investCSV["service_code"]==401]["owner_name"].unique()
selfInspectOwners.shape

#check list
investCSV.loc[investCSV["owner_name"].isin(selfInspectOwners)].head()

```

```python
selfInspectOwners[0:10]
```

```python
# make function that sees if the first score recorded is better or worse than the last recorded score
def getCompare(x):
    x = pd.Series(x)
    # comparisons will take the first valid score and last valid score; by convention, only 1 score will be deemed not improved
    firstScore = x[x.first_valid_index()]
    lastScore = x[x.last_valid_index()]
    return 1*(lastScore > firstScore) #return 1 if higher, 0 if lower

#Q5 -- % of owners who improved with self inspection
#need to aggregate scores by each owner
aggByOwner = investCSV.loc[investCSV["owner_name"].isin(selfInspectOwners)].groupby(["owner_name","Years"])["score"].mean()
aggByOwner = pd.DataFrame(aggByOwner.reset_index())

#pivot the data such that the scores may be compared using the function above
ptOwners = pd.pivot_table(aggByOwner,values=["score"],index=["owner_name"],columns=["Years"])
ptOwners = pd.DataFrame(ptOwners)

ptOwners.head()
```

```python
# check if answers make sense
ptOwners["score"].apply(lambda x: getCompare(x),axis=1).head(5)
```

```python
# percentage improvements with number of self inspecting owners as denominator
print(np.sum(ptOwners["score"].apply(lambda x: getCompare(x),axis=1)))
print(pd.to_numeric(selfInspectOwners.shape)[0])
np.sum(ptOwners["score"].apply(lambda x: getCompare(x),axis=1))/pd.to_numeric(selfInspectOwners.shape)[0]
```
